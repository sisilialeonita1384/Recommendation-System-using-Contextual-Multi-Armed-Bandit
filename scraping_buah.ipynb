CODE TERBARU YG ADA UKURAN DIAMETER

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import csv
import time
import os
import re
from datetime import datetime
import logging

class FatSecretBuahScraper:
    def __init__(self, output_dir="fatsecret_buah_data", delay=2.0):
        self.session = requests.Session()
        self.session.headers.update({
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        })
        
        self.base_url = 'https://www.fatsecret.co.id'
        self.output_dir = output_dir
        self.delay = delay
        os.makedirs(output_dir, exist_ok=True)
        
        logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
        self.logger = logging.getLogger(__name__)
        
        self.fruits_data = []
        self.scraped_urls = set()
        self.processed_variants = set()

    def has_proper_unit(self, text):
        """Cek apakah teks memiliki satuan yang valid - DIPERBAIKI"""
        if not text:
            return False
        
        text_lower = text.lower()
        
        # TOLAK LANGSUNG yang mengandung kata-kata terlarang
        forbidden_words = ['irisan', 'mangkok', 'sdm', 'sendok', 'potong', 'slice', 'cup', 
                          'tbsp', 'tsp', 'sendok makan', 'sendok teh', 'potongan', 'lumatan', 
                          'sdt', 'porsi']
        for word in forbidden_words:
            if word in text_lower:
                return False
        
        # TERIMA jika mengandung pola "sekitar X per Y gram" - DIPERBAIKI POLANYA
        weight_patterns = [
            r'(?:sekitar|kira-?kira)\s*\d+[^)]*per\s*\d+\s*g',  # Pattern lebih fleksibel
            r'\([^)]*(?:sekitar|kira-?kira)[^)]*per[^)]*g[^)]*\)',  # Pattern dalam kurung
        ]
        
        for pattern in weight_patterns:
            if re.search(pattern, text, re.IGNORECASE):
                return True
        
        # TERIMA jika mengandung satuan yang valid
        unit_patterns = [
            r'\d+\s*g\b', r'\d+\s*gram\b', r'\d+\s*kg\b', r'\d+\s*ml\b',
            r'\d+\s*biji\b', r'\d+\s*buah\b', r'\d+\s*ons\b',
            r'\d+\s*apel\b', r'\d+\s*jeruk\b', r'\d+\s*pisang\b',
            r'\d+\s*pepaya\b', r'\d+\s*mangga\b', r'\d+\s*pear\b',
            r'\d+\s*melon\b', r'\d+\s*semangka\b', r'\d+\s*stroberi\b',
            r'\d+\s*nanas\b', r'\d+\s*alpukat\b'
        ]
        
        for pattern in unit_patterns:
            if re.search(pattern, text, re.IGNORECASE):
                return True
        
        # TERIMA jika mengandung diameter (SEMUA UKURAN: kecil, sedang, besar)
        if re.search(r'diameter\s*\d+', text_lower):
            return True
        
        return False

    def get_clean_variant_name(self, text):
        """Bersihkan nama varian - DIPERBAIKI untuk menangani semua ukuran diameter"""
        if not self.has_proper_unit(text):
            return None
            
        clean_text = re.sub(r'Informasi Gizi.*', '', text, flags=re.DOTALL | re.IGNORECASE)
        clean_text = re.sub(r'Database.*', '', clean_text, flags=re.DOTALL | re.IGNORECASE)
        clean_text = re.sub(r'Kalori dalam.*', '', clean_text, flags=re.DOTALL | re.IGNORECASE)
        clean_text = re.sub(r'Energi.*', '', clean_text, flags=re.DOTALL | re.IGNORECASE)
        clean_text = re.sub(r'Lemak.*', '', clean_text, flags=re.DOTALL | re.IGNORECASE)
        clean_text = re.sub(r'Terakhir Diperbarui.*', '', clean_text, flags=re.DOTALL | re.IGNORECASE)
        
        # POLA 1: Ukuran dengan diameter dan berat (SEMUA UKURAN: kecil, sedang, besar)
        # Contoh: "1 kecil (diameter 6,5 cm) (kira-kira 4 per 450 g)"
        # Contoh: "1 sedang (diameter 7 cm) (kira-kira 3 per 450 g)"  
        # Contoh: "1 besar (diameter 8,5 cm) (sekitar 2 per 450 g)"
        full_pattern = r'(\d+\s+(?:kecil|sedang|besar)\s*\([^)]*diameter[^)]*\)\s*\([^)]*(?:sekitar|kira-?kira)[^)]*per[^)]*g[^)]*\))'
        match = re.search(full_pattern, clean_text, re.IGNORECASE)
        if match:
            return match.group(1).strip()
            
        # POLA 2: Ukuran dengan berat tanpa diameter (SEMUA UKURAN)
        weight_pattern = r'(\d+\s+(?:kecil|sedang|besar)\s*\([^)]*(?:sekitar|kira-?kira)[^)]*per[^)]*g[^)]*\))'
        match = re.search(weight_pattern, clean_text, re.IGNORECASE)
        if match:
            return match.group(1).strip()
        
        # POLA 3: Satuan dasar
        unit_match = re.search(r'(\d+\s*(?:g|gram|kg|ml|buah|biji|ons|apel|jeruk|pisang|pepaya|mangga|pear|melon|semangka|stroberi|nanas|alpukat))', clean_text, re.IGNORECASE)
        if unit_match:
            return unit_match.group(1).strip()
        
        # POLA 4: Yang hanya diameter (SEMUA UKURAN)
        diameter_pattern = r'(\d+\s+(?:kecil|sedang|besar)\s*\([^)]*diameter[^)]*\))'
        match = re.search(diameter_pattern, clean_text, re.IGNORECASE)
        if match:
            return match.group(1).strip()
        
        return None

    def extract_actual_food_name_from_page(self, soup, variant_url):
        """Extract nama makanan yang benar dari halaman individual"""
        try:
            # Method 1: Cari di title
            title = soup.find('title')
            if title:
                title_text = title.get_text().strip()
                if "Kalori dalam" in title_text:
                    food_name = re.sub(r'Kalori dalam\s*', '', title_text, flags=re.IGNORECASE)
                    food_name = re.sub(r'\s*-.*', '', food_name)
                    food_name = re.sub(r'\s*dan Fakta Gizi.*', '', food_name, flags=re.IGNORECASE)
                    
                    # Bersihkan nama dari informasi ukuran porsi
                    food_name = re.sub(r'\s*\([^)]*(?:\d+|kecil|sedang|besar|diameter)[^)]*\)', '', food_name, flags=re.IGNORECASE)
                    food_name = re.sub(r'\s*\d+\s*(?:g|gram|kg|ml|buah|biji)', '', food_name, flags=re.IGNORECASE)
                    
                    return food_name.strip()
            
            # Method 2: Cari di heading h1
            h1 = soup.find('h1')
            if h1:
                h1_text = h1.get_text().strip()
                clean_name = re.sub(r'\s*Fakta Gizi.*', '', h1_text, flags=re.IGNORECASE)
                clean_name = re.sub(r'\s*Informasi Gizi.*', '', clean_name, flags=re.IGNORECASE)
                
                # Bersihkan nama dari informasi ukuran porsi
                clean_name = re.sub(r'\s*\([^)]*(?:\d+|kecil|sedang|besar|diameter)[^)]*\)', '', clean_name, flags=re.IGNORECASE)
                clean_name = re.sub(r'\s*\d+\s*(?:g|gram|kg|ml|buah|biji)', '', clean_name, flags=re.IGNORECASE)
                
                return clean_name.strip()
            
            # Method 3: Ekstrak dari URL
            url_pattern = r'/kalori-gizi/[^/]+/([^/?]+)'
            match = re.search(url_pattern, variant_url)
            if match:
                food_name = match.group(1).replace('-', ' ').strip()
                try:
                    from urllib.parse import unquote
                    food_name = unquote(food_name)
                except:
                    pass
                return food_name.title()
                
        except Exception as e:
            self.logger.error(f"Error extracting food name: {e}")
        
        return None

    def extract_nutrition_data_properly(self, soup):
        """Extract data nutrisi dari struktur FatSecret"""
        nutrition_data = {}
        
        try:
            page_text = soup.get_text()
            
            # Extract kalori
            energy_patterns = [
                r'Energi\s*(\d+[.,]?\d*)\s*kj\s*(\d+[.,]?\d*)\s*kkal',
                r'(\d+[.,]?\d*)\s*kkal',
                r'Kal\s*(\d+[.,]?\d*)',
                r'Terdapat\s*(\d+[.,]?\d*)\s*kalori'
            ]
            
            for pattern in energy_patterns:
                match = re.search(pattern, page_text, re.IGNORECASE)
                if match:
                    if len(match.groups()) >= 2:
                        nutrition_data['kalori'] = f"{match.group(2)} kkal"
                    else:
                        nutrition_data['kalori'] = f"{match.group(1)} kkal"
                    break
            
            # Extract nutrisi lainnya
            patterns = {
                'protein': [r'Protein\s*(\d+[.,]?\d*)\s*g', 'g'],
                'karbohidrat': [r'Karbohidrat\s*(\d+[.,]?\d*)\s*g', 'g'],
                'lemak': [r'Lemak\s*(\d+[.,]?\d*)\s*g', 'g'],
                'lemak_jenuh': [r'Lemak Jenuh\s*(\d+[.,]?\d*)\s*g', 'g'],
                'serat': [r'Serat\s*(\d+[.,]?\d*)\s*g', 'g'],
                'gula': [r'Gula\s*(\d+[.,]?\d*)\s*g', 'g'],
                'sodium': [r'Sodium\s*(\d+[.,]?\d*)\s*mg', 'mg'],
                'kalium': [r'Kalium\s*(\d+[.,]?\d*)\s*mg', 'mg'],
                'kolesterol': [r'Kolesterol\s*(\d+[.,]?\d*)\s*mg', 'mg']
            }
            
            for key, (pattern, unit) in patterns.items():
                match = re.search(pattern, page_text, re.IGNORECASE)
                if match:
                    value = match.group(1).replace(',', '.')
                    nutrition_data[key] = f"{value} {unit}"
            
        except Exception as e:
            self.logger.error(f"Error extract nutrisi: {e}")
        
        return nutrition_data

    def scrape_fatsecret_specific(self):
        """Scrape khusus untuk struktur FatSecret - DIPERBAIKI untuk mengambil SEMUA varian"""
        category_url = f'{self.base_url}/kalori-gizi/grup/buah'
        
        try:
            response = self.session.get(category_url)
            if response.status_code != 200:
                self.logger.error("Gagal mengakses halaman kategori buah")
                return []
            
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Cari semua link makanan buah - DIPERBAIKI
            fruit_links = []
            
            # Method 1: Cari semua link yang mengarah ke halaman makanan spesifik
            all_links = soup.find_all('a', href=re.compile(r'/kalori-gizi/'))
            for link in all_links:
                href = link.get('href')
                text = link.get_text().strip()
                
                # Filter hanya item makanan spesifik (bukan kategori)
                if (href and text and len(text) > 2 and
                    href.startswith('/kalori-gizi/') and
                    not href.endswith('/grup/') and
                    'grup' not in href and
                    not any(word in text.lower() for word in ['kategori', 'grup', 'lihat semua', 'selengkapnya', 'semua', 'makanan'])):
                    
                    full_url = urljoin(self.base_url, href)
                    fruit_links.append((full_url, text))
            
            # Hapus duplikat
            fruit_links = list(dict.fromkeys(fruit_links))
            self.logger.info(f"Ditemukan {len(fruit_links)} item buah potensial")
            
            # Scrape setiap buah
            for i, (fruit_url, fruit_name) in enumerate(fruit_links, 1):
                try:
                    self.logger.info(f"[{i}/{len(fruit_links)}] Memproses: {fruit_name}")
                    
                    if fruit_url in self.scraped_urls:
                        continue
                        
                    fruit_response = self.session.get(fruit_url)
                    if fruit_response.status_code != 200:
                        self.logger.warning(f"Gagal mengakses: {fruit_url}")
                        continue
                    
                    fruit_soup = BeautifulSoup(fruit_response.content, 'html.parser')
                    self.scraped_urls.add(fruit_url)
                    
                    # Cari SEMUA varian ukuran porsi - DIPERBAIKI
                    variants = []
                    
                    # Method 1: Cari semua link portionid (PRIORITAS TERTINGGI)
                    portion_links = fruit_soup.find_all('a', href=re.compile(r'portionid='))
                    for link in portion_links:
                        variant_text = link.get_text().strip()
                        if variant_text:
                            clean_name = self.get_clean_variant_name(variant_text)
                            if clean_name:
                                full_url = urljoin(self.base_url, link.get('href'))
                                variants.append((full_url, clean_name))
                    
                    # Method 2: Cari di tabel common servings
                    common_servings = fruit_soup.find_all('div', class_=re.compile(r'serving|portion', re.I))
                    for serving in common_servings:
                        serving_text = serving.get_text().strip()
                        if serving_text:
                            clean_name = self.get_clean_variant_name(serving_text)
                            if clean_name:
                                variants.append((fruit_url, clean_name))
                    
                    # Method 3: Cari di tabel data
                    tables = fruit_soup.find_all('table')
                    for table in tables:
                        rows = table.find_all('tr')
                        for row in rows:
                            cells = row.find_all(['td', 'th'])
                            if cells and len(cells) > 0:
                                variant_text = cells[0].get_text().strip()
                                if variant_text:
                                    clean_name = self.get_clean_variant_name(variant_text)
                                    if clean_name:
                                        variants.append((fruit_url, clean_name))
                    
                    # Hapus duplikat dalam varian untuk buah yang sama
                    unique_variants = []
                    seen_variants = set()
                    for url, variant in variants:
                        if variant not in seen_variants:
                            unique_variants.append((url, variant))
                            seen_variants.add(variant)
                    
                    self.logger.info(f"   Ditemukan {len(unique_variants)} varian unik untuk {fruit_name}")
                    
                    # Scrape setiap varian yang valid
                    for j, (variant_url, variant_name) in enumerate(unique_variants, 1):
                        try:
                            # Buat identifier unik untuk varian
                            variant_id = f"{fruit_name}_{variant_name}"
                            if variant_id in self.processed_variants:
                                self.logger.info(f"   Varian {j}/{len(unique_variants)}: {variant_name} - ❌ DUPLIKAT, dilewati")
                                continue
                                
                            self.logger.info(f"   Varian {j}/{len(unique_variants)}: {variant_name}")
                            
                            # Akses halaman varian untuk mendapatkan data yang akurat
                            if variant_url != fruit_url:
                                variant_response = self.session.get(variant_url)
                                if variant_response.status_code != 200:
                                    continue
                                variant_soup = BeautifulSoup(variant_response.content, 'html.parser')
                                actual_food_name = self.extract_actual_food_name_from_page(variant_soup, variant_url) or fruit_name
                                nutrition_data = self.extract_nutrition_data_properly(variant_soup)
                            else:
                                actual_food_name = self.extract_actual_food_name_from_page(fruit_soup, fruit_url) or fruit_name
                                nutrition_data = self.extract_nutrition_data_properly(fruit_soup)
                            
                            if nutrition_data:
                                fruit_data = {
                                    'nama_buah': actual_food_name,
                                    'ukuran_porsi': variant_name,
                                    'url': variant_url,
                                    'kalori': nutrition_data.get('kalori', 'Tidak ditemukan'),
                                    'protein': nutrition_data.get('protein', ''),
                                    'karbohidrat': nutrition_data.get('karbohidrat', ''),
                                    'lemak': nutrition_data.get('lemak', ''),
                                    'lemak_jenuh': nutrition_data.get('lemak_jenuh', ''),
                                    'serat': nutrition_data.get('serat', ''),
                                    'gula': nutrition_data.get('gula', ''),
                                    'sodium': nutrition_data.get('sodium', ''),
                                    'kalium': nutrition_data.get('kalium', ''),
                                    'kolesterol': nutrition_data.get('kolesterol', '')
                                }
                                
                                self.fruits_data.append(fruit_data)
                                self.processed_variants.add(variant_id)
                                self.logger.info(f"      ✅ Berhasil - {actual_food_name} - Kalori: {nutrition_data.get('kalori', 'N/A')}")
                            
                            time.sleep(self.delay / 2)
                            
                        except Exception as e:
                            self.logger.error(f"      Error scraping varian: {e}")
                    
                    time.sleep(self.delay)
                    
                except Exception as e:
                    self.logger.error(f"Error processing {fruit_name}: {e}")
            
        except Exception as e:
            self.logger.error(f"Error scraping kategori: {e}")
        
        return self.fruits_data

    def save_to_csv(self):
        """Save ke CSV"""
        if not self.fruits_data:
            return None
        
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        csv_file = f'{self.output_dir}/buah_corrected_{timestamp}.csv'
        
        fieldnames = [
            'nama_buah', 'ukuran_porsi', 'url',
            'kalori', 'protein', 'karbohidrat', 'lemak', 'lemak_jenuh',
            'serat', 'gula', 'sodium', 'kalium', 'kolesterol'
        ]
        
        with open(csv_file, 'w', newline='', encoding='utf-8') as f:
            writer = csv.DictWriter(f, fieldnames=fieldnames)
            writer.writeheader()
            
            for fruit in self.fruits_data:
                writer.writerow(fruit)
        
        # Tampilkan preview
        print("\n" + "="*80)
        print("PREVIEW DATA YANG DIHASILKAN:")
        print("="*80)
        with open(csv_file, 'r', encoding='utf-8') as f:
            lines = f.readlines()
            for line in lines[:10]:
                print(line.strip())
        
        return csv_file

# Jalankan
if __name__ == "__main__":
    scraper = FatSecretBuahScraper(delay=2)
    
    results = scraper.scrape_fatsecret_specific()
    
    if results:
        csv_file = scraper.save_to_csv()
        print(f"\n🎉 SCRAPING SELESAI!")
        print(f"📊 Total data: {len(results)}")
        print(f"📁 File: {csv_file}")
        
        # Tampilkan statistik
        print("\n📈 STATISTIK UKURAN PORSI:")
        
        weight_info_count = sum(1 for d in results if any(word in d['ukuran_porsi'].lower() for word in ['sekitar', 'kira']))
        diameter_count = sum(1 for d in results if 'diameter' in d['ukuran_porsi'].lower())
        basic_unit_count = len(results) - weight_info_count - diameter_count
        
        print(f"   Dengan info berat: {weight_info_count}")
        print(f"   Dengan diameter: {diameter_count}")
        print(f"   Satuan dasar: {basic_unit_count}")
        print(f"   Total varian: {len(results)}")
        
        # Tampilkan contoh data berbagai jenis
        print("\n📋 CONTOH DATA BERBAGAI JENIS UKURAN PORSI:")
        
        # Contoh dengan info berat
        weight_examples = [d for d in results if any(word in d['ukuran_porsi'].lower() for word in ['sekitar', 'kira'])]
        for example in weight_examples[:3]:
            print(f"\n🔸 {example['nama_buah']} - {example['ukuran_porsi']}")
            print(f"   Kalori: {example.get('kalori', 'N/A')}")
        
        # Contoh dengan diameter
        diameter_examples = [d for d in results if 'diameter' in d['ukuran_porsi'].lower()]
        for example in diameter_examples[:3]:
            print(f"\n🔹 {example['nama_buah']} - {example['ukuran_porsi']}")
            print(f"   Kalori: {example.get('kalori', 'N/A')}")
        
        # Contoh dengan satuan dasar
        basic_examples = [d for d in results if not any(word in d['ukuran_porsi'].lower() for word in ['sekitar', 'kira', 'diameter'])]
        for example in basic_examples[:3]:
            print(f"\n🔸 {example['nama_buah']} - {example['ukuran_porsi']}")
            print(f"   Kalori: {example.get('kalori', 'N/A')}")
            
    else:
        print("❌ Tidak ada data yang berhasil di-scrape")
```
